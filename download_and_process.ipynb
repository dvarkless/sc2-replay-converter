```python
processor = ReplayProcess(
    SECRETS,
    DATABASE_CONFIG,
    GAME_INFO_FILE,
    jupyter=True
)
```

```python
replay_filter = ReplayFilter()
replay_filter.is_1v1 = True
replay_filter.game_len = [1920, 38400]
replay_filter.time_played = datetime(2021, 1, 1)
```

```python
processor.process_replays(REPLAY_DIR, filt=replay_filter)
```

```python
MINS_PER_SAMPLE = 4
PRED_STEP = 1
MIN_LEAGUE = 3
```

```python
r_pairs = product("ZTP", repeat=2)
matchups = ["v".join((r1, r2)) for r1, r2 in r_pairs]
composer = PipelineComposer("ZvZ", tick_step=32)
```

```python
for matchup in matchups:
    composer.change_matchup(matchup)
    comp_pipeline = composer.get_compositon(MINS_PER_SAMPLE, PRED_STEP, MIN_LEAGUE)
    comp_pipeline.run()
```

```python
for matchup in matchups:
    composer.change_matchup(matchup)
    winprob_pipeline = composer.get_win_probability(MINS_PER_SAMPLE, PRED_STEP, MIN_LEAGUE)
    winprob_pipeline.run()
```

```python
for matchup in matchups:
    composer.change_matchup(matchup)
    enemycomp_pipeline = composer.get_enemy_composition(MINS_PER_SAMPLE, PRED_STEP, MIN_LEAGUE)
    enemycomp_pipeline.run()
```
